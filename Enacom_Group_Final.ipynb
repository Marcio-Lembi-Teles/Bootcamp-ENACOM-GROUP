Devido os anteriores n√£o conterem essa informa√ß√£o e, a mesma ser uma coluna de dado tipo "texto", resolvi exclu√≠-la desses dois dataframes para facilitar a uni√£o com os demais dataframes para obter apenas um conjunto de dados para realiza√ß√£o das respectivas an√°lises.

```python
# Lista de todos os DataFrames
todos_dados = [dados_2000, dados_2001, dados_2002, dados_2003, dados_2004,
               dados_2005, dados_2006, dados_2007, dados_2008, dados_2009,
               dados_2010, dados_2011, dados_2012, dados_2013, dados_2014,
               dados_2015, dados_2016, dados_2017, dados_2018, dados_2019,
               dados_2020]
```

```python
# Importar bibliotecas
from statsmodels.tsa.arima.model import ARIMA

#USO DO MODELO S√âRIES TEMPORAIS - ARIMA para an√°lises de previs√£o

# Carregar dados de treinamento
dados_treino = pd.read_csv('dados_treino.csv', parse_dates=['Ano-M√™s'], index_col='Ano-M√™s')

# Visualizar os primeiros registros dos dados de treinamento
print("Dados de Treinamento:")
print(dados_treino.head())

# Carregar dados de teste
dados_teste = pd.read_csv('dados_teste.csv', parse_dates=['Ano-M√™s'], index_col='Ano-M√™s')

# Visualizar os primeiros registros dos dados de teste
print("\nDados de Teste:")
print(dados_teste.head())

# Plotar a s√©rie temporal de treinamento
plt.figure(figsize=(10, 6))
plt.plot(dados_treino, label='Treinamento')
plt.title('S√©rie Temporal de Treinamento')
plt.xlabel('Ano-M√™s')
plt.ylabel('Total Mensal')
plt.legend()
plt.show()

# Ajustar o modelo ARIMA
modelo = ARIMA(dados_treino, order=(5,1,0))  # Ordem (p, d, q) do modelo ARIMA
resultado = modelo.fit()

# Fazer previs√µes
previsoes = resultado.predict(start=len(dados_treino), end=len(dados_treino) + len(dados_teste) - 1, typ='levels')

# Plotar previs√µes
plt.figure(figsize=(10, 6))
plt.plot(dados_treino, label='Treinamento')
plt.plot(dados_teste, label='Teste')
plt.plot(previsoes.index, previsoes, label='Previs√µes', color='red')
plt.title('Previs√µes com Modelo ARIMA')
plt.xlabel('Ano-M√™s')
plt.ylabel('Total Mensal')
plt.legend()
plt.show()
```
